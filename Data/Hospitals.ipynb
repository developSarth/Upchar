```python
import pandas as pd

import random
cf = pd.read_csv('final_master_clinics_india.csv', encoding='utf-8')

# Drop unnecessary columns
cf.drop(['lat','lon','postcode','id','amenity','website','street','type','district','locality'], 
        axis=1, inplace=True, errors='ignore')

# Keep only rows where name, lat, and full address exist
cf.dropna(subset=['name', 'address_full'], inplace=True)
cf.rename(columns={'address_full': 'Address'}, inplace=True)
cf.rename(columns={'name': 'Clinic Name'}, inplace=True)
cf.rename(columns={'opening_hours': 'Opening Hours'}, inplace=True)
cf.rename(columns={'speciality': 'Speciality'}, inplace=True)
cf.rename(columns={'phone': 'Contact'}, inplace=True)
cf.rename(columns={'city': 'City'}, inplace=True)
cf.rename(columns={'state': 'State'}, inplace=True)
# cf.drop('cf['healthcare']'='hospital')
cf = cf[cf['healthcare'].str.lower() != 'hospital']
cf.drop('healthcare', axis=1, inplace=True, errors='ignore')
cf['Opening Hours'] = cf['Opening Hours'].fillna('10 A.M. to 2 P.M. & 5 P.M. to 10 P.M.')
mask = cf['Contact'].isna()
cf.drop_duplicates(inplace=True)
# Generate random unique numbers for NA rows
# cf['Contact'] = cf['Contact'].fillna(pd.Series([random.randint(1000000000, 9999999999) for _ in range(len(cf))]))
cf.to_csv('final_master_clinics_india.csv', index=False, encoding='utf-8')


```
